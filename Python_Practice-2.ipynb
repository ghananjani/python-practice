```python
#float numbers with range

for i in range(0,1.1,2):
  print(i)
```

```python
#function for range with floats

def frange(start,stop,step):
  i = start
  while i < stop:
    yield i
    i+= step

for i in frange(0.1,1.0,0.2):
  print(i)
```

```python
#concatenation of 2 range function

from itertools import chain
res= chain (range(10), range(10,21))
for i in res:
  print(i, end = ",")
```
