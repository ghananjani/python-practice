```python
def test_fib(n):
  a,b = 0,1
  for i in range(n):
    yield a
    a,b = b, a+b
```

```python
test_fib(10)
```

```python
for i in test_fib(10):
  print(i)
```

```python
def fibonacci():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

# Create the iterator
fib = fibonacci()

# Print the first 10 numbers of the Fibonacci sequence
for i in range(10):
    print(next(fib))
```

```python
def count_test(n):
  count = 1
  while count <= n:
    yield count
    count = count + 1
```

```python
c = count_test(5)
```

```python
for i in c:
  print(i)
```
